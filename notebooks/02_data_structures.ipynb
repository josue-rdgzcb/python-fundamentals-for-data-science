# Estructuras de datos

## Objetivos

- Comprender qué son las estructuras de datos y por qué son fundamentales en programación.
- Identificar las principales estructuras de datos integradas en Python.
- Crear y manipular listas, tuplas, conjuntos y diccionarios.
- Acceder, modificar y organizar datos dentro de estas estructuras.
- Elegir la estructura de datos adecuada según el tipo de problema.

## Conceptos clave

***Estructura de datos:***  
Forma de organizar y almacenar información para que pueda ser accedida y manipulada de manera eficiente.

***Colección:***  
Conjunto de múltiples valores almacenados dentro de una misma estructura de datos.

***Lista (list):***  
Estructura de datos ordenada y mutable que permite almacenar múltiples elementos en una sola variable. Los elementos pueden modificarse, agregarse o eliminarse. Definidas entre corchetes [], pueden contener tipos de datos mixtos (números, cadenas, booleanos).

***Tupla (tuple):***  
Estructura de datos ordenada pero inmutable. Una vez creada, sus elementos no pueden modificarse. Se definen utlizando paréntesis ().

***Mutabilidad:***  
Propiedad de una estructura de datos que permite modificar su contenido después de haber sido creada.

***Diccionario (dict):***  
Estructura de datos que almacena pares de clave–valor, permitiendo acceder a los valores mediante una clave única.

***Clave (key):***  
Identificador único dentro de un diccionario que se utiliza para acceder a un valor asociado.

***Valor (value):***  
Dato almacenado en un diccionario que está asociado a una clave específica.

***Conjunto (set):***  
Colección no ordenada de elementos únicos. No permite valores duplicados y se utiliza comúnmente para operaciones de teoría de conjuntos.

***Índice:***  
Posición de un elemento dentro de una secuencia ordenada. En Python los índices comienzan en 0.

***Slicing:***  
Técnica que permite acceder a subconjuntos de elementos dentro de una secuencia utilizando rangos de índices.

***Elemento único:***  
Valor que aparece una sola vez dentro de una colección, sin duplicados.
