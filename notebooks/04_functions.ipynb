# Funciones

## Objetivos

- Comprender qué es una función y por qué es útil en programación.
- Definir funciones propias utilizando def.
- Utilizar parámetros para pasar información a una función.
- Devolver resultados mediante la sentencia return.
- Reutilizar funciones para estructurar y simplificar programas.
- Escribir funciones claras, reutilizables y fáciles de mantener.

## Conceptos clave

***Función:***  
Bloque de código reutilizable que realiza una tarea específica y que puede ejecutarse cuando es llamado. Una función se crea utilizando la palabra clave def, especificando su nombre, parámetros y bloque de código.

***Llamada a función:***  
Ejecución de una función previamente definida utilizando su nombre.

***Parámetro:***  
Variable definida en la declaración de una función que recibe un valor cuando la función es llamada.

***Argumento:***  
Valor real que se pasa a una función cuando es llamada.

***Valor de retorno (return):***  
Resultado que una función devuelve al finalizar su ejecución mediante la sentencia return.

***Alcance de variables (scope):***  
Región del programa donde una variable es accesible.

***Variable local:***  
Variable definida dentro de una función y accesible únicamente dentro de ella.

***Variable global:***  
Variable definida fuera de una función y accesible desde diferentes partes del programa.

***Docstring:***  
Cadena de texto utilizada para documentar una función, explicando su propósito, parámetros y valores de retorno.
