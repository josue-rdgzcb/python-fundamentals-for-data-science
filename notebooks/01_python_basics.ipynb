## Conceptos clave

***Python:***   
Lenguaje de programación de alto nivel e interpretado, conocido por su sintaxis clara y legible. Es ampliamente utilizado en ciencia de datos, inteligencia artificial, desarrollo web y automatización.

***Intérprete:***  
Programa que ejecuta el código Python línea por línea, traduciendo las instrucciones a acciones que la computadora puede realizar.

***Variable:***  
Nombre que hace referencia a un valor almacenado en memoria. Permite guardar datos y reutilizarlos dentro de un programa.

***Valor:***  
Dato específico almacenado en una variable, como un número, texto o valor lógico.

***Tipo de dato:***  
Clasificación que define la naturaleza de un valor y las operaciones que pueden realizarse con él.

***Entero (int):***  
Número sin parte decimal, positivo o negativo.  
Ejemplo: 5, -12, 100.

***Número flotante (float):***  
Número que contiene parte decimal.  
Ejemplo: 3.14, 0.5, -7.2.

***Cadena de texto (str):***  
Secuencia de caracteres utilizada para representar texto. Se escribe entre comillas simples o dobles.  
Ejemplo: "hola".

***Booleano (bool):***  
Tipo de dato lógico que solo puede tomar dos valores: True (verdadero) o False (falso).

***Operador:***  
Símbolo que realiza una operación sobre uno o más valores.  
Ejemplos: +, -, *, /, ==, >.

***Expresión:***  
Combinación de valores, variables y operadores que Python evalúa para producir un resultado.

***Salida (output):***  
Información que un programa muestra al usuario, generalmente utilizando la función print().

***Comentario:***  
Texto dentro del código que no es ejecutado por Python y sirve para explicar o documentar el programa. Se escribe utilizando #
