# Control de flujo

## Objetivos

- Comprender cómo los programas pueden tomar decisiones y repetir acciones.
- Utilizar estructuras condicionales para ejecutar diferentes bloques de código según ciertas condiciones.
- Implementar bucles para automatizar tareas repetitivas.
- Controlar el comportamiento de los bucles mediante instrucciones específicas.
- Aplicar estructuras de control para resolver problemas básicos de programación.

## Conceptos clave

***Flujo de control:***  
Orden en el que se ejecutan las instrucciones dentro de un programa.

***Estructura condicional:***  
Mecanismo que permite ejecutar diferentes bloques de código dependiendo de si una condición es verdadera o falsa.

***Condición:***  
Expresión que se evalúa como True o False y determina qué camino sigue el programa.

***Sentencia if:***  
Estructura que ejecuta un bloque de código si una condición es verdadera.

***Sentencia elif:***  
Permite evaluar condiciones adicionales después de un if cuando la condición inicial no se cumple.

***Sentencia else:***  
Bloque que se ejecuta cuando ninguna de las condiciones anteriores es verdadera.

***Bucle (loop):***  
Estructura que permite repetir un bloque de código varias veces.

***Bucle for:***  
Tipo de bucle que itera sobre una secuencia o colección de elementos.

***Bucle while:***  
Bucle que ejecuta un bloque de código mientras una condición sea verdadera.

***Iteración:***  
Cada repetición individual de un bucle.

***Control de bucles:***  
Instrucciones que permiten modificar el comportamiento normal de un bucle.

***Sentencia break:***  
Interrumpe inmediatamente la ejecución de un bucle.

***Sentencia continue:***  
Omite el resto de la iteración actual y pasa directamente a la siguiente.

***Anidación:***  
Situación en la que una estructura de control (como un if o un bucle) se encuentra dentro de otra.
