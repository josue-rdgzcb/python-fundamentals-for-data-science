# Módulos y paquetes

## Objetivos

Al finalizar este notebook deberías ser capaz de:

- Comprender qué son los módulos y paquetes en Python.
- Importar y utilizar módulos de la biblioteca estándar.
- Importar funciones, clases o variables específicas desde un módulo.
- Utilizar alias para simplificar nombres de módulos.
- Organizar código de manera modular para mejorar la reutilización y mantenimiento.

## Conceptos clave

***Módulo:***  
Archivo de Python (.py) que contiene código reutilizable, como funciones, clases o variables, que puede ser importado en otros programas.

***Paquete:***  
Colección de módulos organizados en un directorio que permite estructurar proyectos más grandes y mantener el código ordenado.

***Importación:***  
Proceso de acceder a funciones, clases o variables definidas en otro módulo utilizando la palabra clave import.

***Biblioteca estándar:***  
Conjunto de módulos incluidos por defecto en Python que proporcionan herramientas para tareas comunes, como cálculos matemáticos, manejo de archivos y generación de números aleatorios.

***Sentencia import:***  
Instrucción que permite cargar un módulo completo dentro de un programa.

***Importación específica:***  
Forma de importar elementos particulares de un módulo utilizando from modulo import objeto.

***Alias de importación:***  
Nombre alternativo que se asigna a un módulo al importarlo para simplificar su uso, por ejemplo *import numpy as np*.

***Dependencia:***  
Módulo o paquete externo del cual depende un programa para funcionar correctamente.

***Modularidad:***  
Principio de diseño que consiste en dividir un programa en componentes independientes y reutilizables.
